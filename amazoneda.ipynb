```python
import pandas as pd
import numpy as np
df = pd.read_csv('D:\\Python\\Projects\\Amazon EDA\\amazon.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['discounted_price'] = df['discounted_price'].astype(str).str.replace("₹", "", regex=False).str.replace(",", "", regex=False)
df['actual_price'] = df['actual_price'].astype(str).str.replace("₹", "", regex=False).str.replace(",", "", regex=False)

df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
```

```python
df['rating'].unique()
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df["rating_count"].unique()
df["rating_count"] = (
    df["rating_count"]
    .astype(str)
    .str.replace(",", "", regex=False)
)

df["rating_count"] = pd.to_numeric(
    df["rating_count"], errors="coerce"
)
```

```python
df.info()
```

```python
df[['rating', 'rating_count']].isna().sum()
```

```python
df["main_category"] = df["category"].str.split("|").str[0]
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#univariate Analysis
```

```python
sns.histplot(df['actual_price'], bins=30,kde=True)
plt.title("Price Distribution")
plt.show()
```

```python
sns.countplot(y='main_category', data=df, order=df['main_category'].value_counts().index)
plt.title('Product Count by Category')
plt.show()
```

```python
#BiVariate Analysis

plt.figure(figsize=(12,6))
sns.boxplot(x='main_category', y='actual_price', data=df)
plt.xticks(rotation=90)
plt.title('Price Distribution by Category')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['actual_price','rating','rating_count']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
